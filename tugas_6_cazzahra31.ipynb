<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

Jawab:

```python
# code here

# Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
desc = diamonds_data.describe()
display(desc)

all_request = diamonds_data.describe().loc[['mean','std','max']]
display(all_request)

# Berapa nilai rata-rata dari column price
rata_rata = diamonds_data['price'].mean()
print("Rata-rata kolom price:",rata_rata)

# Berapa nilai standar deviasi dari column depth
standar_deviasi = diamonds_data['depth'].std()
print("Standar deviasi kolom depth:",standar_deviasi)

# Berapa nilai maximum dari column carat
max_carat = diamonds_data['carat'].max()
print("nilai maximum kolom carat:",max_carat)
```
