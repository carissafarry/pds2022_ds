```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
df_group = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
df_group
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here

# 1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
import numpy as np

df_transpose = df_group.transpose()
x = np.arange(len(df_transpose))
width = 0.07

fig, ax = plt.subplots(figsize=(17,8))

for i,j in enumerate(df_transpose.columns):
  ax.bar(x + (width*i), df_transpose[j], width=width, label=j)

plt.style.use('ggplot')
plt.title('Mean Sales Video Games By Genre')
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.xticks(x+0.42, labels=df_transpose.index)
plt.legend(loc='best', frameon=1, bbox_to_anchor=(1.06, 1.16))
plt.show
```

Jawab:

// 2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

**Dari keseluruhan data**
- Penjualan video games tertinggi ada pada jenis Platform, yang terjual pada region NA.
- Sedangkan penjualan video games terendah ada pada jenis Adventure, yang terjual pada region lain, selain NA, EU, dan JP.

**Per jenis game**
- Penjualan semua jenis video games kecuali jenis Role-Playing, paling tinggi terjual pada region NA.
- Penjualan video games berjenis Role-Playing, paling tinggi terjual pada region JP.
- Penjualan semua jenis video games kecuali jenis Action, Racing, dan Shooter, paling rendah terjual pada region lainnya.
- Penjualan video games berjenis Action, Racing, dan Shooter, paling rendah terjual pada region JP.

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here

# 1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
fig, ax = plt.subplots(figsize=(16,8))

for i in range(len(df_group.index)) :
  ax.bar(df_transpose.index, df_transpose[df_group.index[i]], bottom=df_transpose[df_group.index[:i]].sum(axis=1), label=df_group.index[i])

plt.style.use('ggplot')
plt.title('Mean Sales Video Games By Genre')
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.legend(loc='best', frameon=1, bbox_to_anchor=(1.06, 1.16))
plt.show
```

Jawab:

// 2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

- Rata-rata keseluruhan penjualan terbanyak, didapat oleh region NA.
- Sedangkan rata-rata keseluruhan penjualan paling sedikit, dijual oleh region lainnya, yaitu selain NA, EU, dan JP.
