<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig, ax = plt.subplots(figsize=(18,6.7), nrows=1, ncols=3, sharex=True, sharey=True)
ax = ax.flatten()
ax[0].plot(x, tan, color='b', label='tanX')
ax[1].plot(x, sin, color='r', label='sinX')
ax[2].plot(x, cos, color='g', label='cosX')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
# bounds : [x0, y0, width, height]
fig, ax = plt.subplots(figsize=(13,8.2))
ax.plot(x, y)

ins1 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins1.plot(x, y5, 'khaki')
ins1.plot(x, y, 'b')
ins1.set_title('Grafik CosX^5')

ins2 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins2.plot(x, y4, 'black')
ins2.plot(x, y, 'b')
ins2.set_title('Grafik CosX^4')

ins3 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3.plot(x, y2, 'r')
ins3.plot(x, y, 'b')
ins3.set_title('Grafik CosX^2')

ins4 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins4.plot(x, y3, 'g')
ins4.plot(x, y, 'b')
ins4.set_title('Grafik CosX^3')

plt.show()
```
